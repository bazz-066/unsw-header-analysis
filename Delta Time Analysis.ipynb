```python
delta_time = benign_set["avg_delta_time_a2b"].append(malicious_set["avg_delta_time_a2b"])
labels = pd.Series(np.zeros(len(benign_set)))
labels = labels.append(pd.Series(np.ones(len(malicious_set))), ignore_index=True)

X = delta_time.values.reshape(-1,1)
Y = labels.values.ravel()
X_meta = metasploit_set["avg_delta_time_a2b"].values.reshape(-1,1)
Y_meta = np.ones(len(metasploit_set)).ravel()

X_scaled = preprocessing.scale(X)
X_meta_scaled = preprocessing.scale(X_meta)
X_scaled
```

```python
def get_scores(Y_true, Y_pred):
    scores = {}
    scores["f1"] = metrics.f1_score(Y_true, Y_pred)
    scores["precision"] = metrics.precision_score(Y_true, Y_pred)
    scores["recall"] = metrics.recall_score(Y_true, Y_pred)
    scores["accuracy"] = metrics.accuracy_score(Y_true, Y_pred)
    scores["tn"], scores["fp"], scores["fn"], scores["tp"] = metrics.confusion_matrix(Y_true, Y_pred, labels=[0,1]).ravel()
    return scores
    
def get_avg_score(scores):
    avg_scores = {}
    for key in scores[0]:
        avg_scores[key] = 0
    
    for score in scores:
        for key, value in score.items():
            avg_scores[key] += value

    for key, value in avg_scores.items():
        avg_scores[key] /= float(len(scores))

    return avg_scores
```
